```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from io import StringIO
import os

# Define the URL of the directory
url = "https://www2.census.gov/econ/bps/County/"

# Send a request to the URL and get the page content
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Find all links on the page
for link in soup.find_all('a'):
    href = link.get('href')
    if href and href.endswith('y.txt'):
        file_url = url + href
        file_name = href.split('/')[-1]

        # Fetch the data from the URL and read it into a DataFrame
        response = requests.get(file_url)
        data = response.text
        
        # Split the data into lines and skip the first two rows
        lines = data.split('\n')[2:]
        cleaned_data = '\n'.join(lines)
        
        # Define your own header
        header = ['FIPS_State_Code', 'FIPS_County_Code', 'Census_Region_Code', 
                  'Census_Division_Code', 'County_Name', 
                  'Estimated_1_Unit_Buildings', 'Estimated_1_Unit_Total_Units', 'Estimated_1_Unit_Total_Value', 
                  'Estimated_2_Unit_Buildings', 'Estimated_2_Unit_Total_Units', 'Estimated_2_Unit_Total_Value', 
                  'Estimated_3to4_Unit_Buildings', 'Estimated_3to4_Unit_Total_Units', 'Estimated_3to4_Unit_Total_Value', 
                  'Estimated_5plus_Unit_Buildings', 'Estimated_5plus_Unit_Total_Units', 'Estimated_5plus_Unit_Total_Value', 
                  'Reported_1_Unit_Buildings', 'Reported_1_Unit_Total_Units', 'Reported_1_Unit_Total_Value',
                  'Reported_2_Unit_Buildings', 'Reported_2_Unit_Total_Units', 'Reported_2_Unit_Total_Value', 
                  'Reported_3to4_Unit_Buildings', 'Reported_3to4_Unit_Total_Units', 'Reported_3to4_Unit_Total_Value', 
                  'Reported_5plus_Unit_Buildings', 'Reported_5plus_Unit_Total_Units', 'Reported_5plus_Unit_Total_Value', 
                 ]  # Replace with your own header names
        
        # Read data into a DataFrame with your defined header
        df = pd.read_csv(StringIO(cleaned_data), delimiter=',', header=None, names=header)

        # Define the path for the output CSV file in the Downloads folder
        csv_file = os.path.join(os.path.expanduser('~/Downloads'), f'{os.path.splitext(file_name)[0]}.csv')

        # Write the DataFrame to a CSV file
        df.to_csv(csv_file, index=False)

        print(f'Converted {file_name} to {os.path.basename(csv_file)}')
```
