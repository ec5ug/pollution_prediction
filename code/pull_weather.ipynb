```python
# Import Meteostat library and dependencies
from datetime import datetime
from meteostat import Point, Hourly

austin = Point(30.2500, -97.7500, 150) # Austin, Texas
albemarle = Point(38.0293, -78.4767, 0) # Albemarle County, Virginia
arlington = Point(38.8783, -77.0687, 0) # Arlington, Virginia
charles_city = Point(37.3312, -77.0615, 0) # CHarles City, Virginia

def grab_data(location, region_name):
    # Set time period
#     start = datetime(1970, 1, 1)
#     end = datetime(2023, 10, 21)
    start = datetime(2018, 1, 1)
    end = datetime(2018, 12, 31)
    # Get hourly data for the specified parameters
    data = Hourly(location, start, end)
    data = data.normalize().interpolate()
    # Extract specific parameters
    parameters = data.fetch()
    cleaned_parameters = parameters.dropna(how='all')
    cleaned_parameters['county'] = region_name
    print(f"{region_name}: start time {cleaned_parameters.index[0]} end time {cleaned_parameters.index[-1]}")
    return cleaned_parameters
```
